# matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False

# 한국어가 적힌 데이터를 시각화 하기 위한 기본 설정

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')

plt.show()
```

# seaborn

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
#데이터 불러오기
penguins = sns.load_dataset("penguins")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
#왼쪽 축 테두리 삭제
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")

#레전드 타이틀 삭제
g.legend.set_title("")
plt.show()
```

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g', hue='sex', color='gray')
sns.stripplot(data=penguins, x='species', y='body_mass_g', hue='sex')

#두 개의 레전드 타이틀 삭제
plt.legend(title = "")
```

```python
#데이터 불러오기
tips = sns.load_dataset("tips")

#특정 행 정렬하기
tips.sort_values('total_bill', inplace=True)
```

```python
#그룹짓기
# group 1: 3~10
# group 2: 10~20
# group 3: 20~

def check_num(x):
    if x <= 10:
        return 'low'
    elif x <= 20:
        return 'middle'
    else:
        return 'high'
```

```python
#그룹 적용하기
tips['total_bill_cat'] = tips['total_bill'].apply(lambda x: check_num(x))
```

```python
sns.barplot(data=tips, x='total_bill_cat', y='tip', color='darkorange')
plt.xlabel('total bill (low, middle, high)')
plt.ylabel('tip (dollors)')
plt.title('tip - total bill')
plt.show()
```

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

```python
df = px.data.gapminder()
```

```python
df['year'].unique()
```

```python
df.query("year == 2007 and continent == 'Europe'", inplace=True)
```

```python
#각 데이터에 있는 행을 순회하기 위해 index를 범위로 잡음 (df를 범위로 잡으면 열이름을 순회한다)
for i in df.index:
    if df.loc[i, 'pop'] < 2.e6:
        df.loc[i, 'country'] = 'Other countries'
```

```python
fig = px.pie(data_frame=df, values='pop', names='country', title='example_title')
fig.show() #fig.write_html('test.html')
```
